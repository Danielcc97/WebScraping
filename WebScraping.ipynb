### Prueba de la librería BeautifulSoup con el fin de conocer la página principal de Books to Scrape

```
from bs4 import BeautifulSoup
import requests
page = requests.get("http://books.toscrape.com")
soup = BeautifulSoup(page.content)
```

```
print(soup.prettify())
```

Obtención de prueba de las categorías existentes en la parte izquierda de la página principal

```

for texto in soup.find_all("div",class_="side_categories"):
  vector = texto.getText().strip()
```

```
vector = vector.replace("\n","")
vector = vector.replace("Books","")
vector
```

```
import re
categorias = re.findall("\w+",vector)
print(categorias)
```

```
categorias = ['Travel', 'Mystery', 'Historical Fiction', 'Sequential Art', 'Classics', 'Philosophy', 'Romance', 'Womens Fiction', 'Fiction', 'Childrens', 'Religion', 'Nonfiction', 'Music', 'Default', 'Science Fiction', 'Sports and Games', 'Add a comment', 'Fantasy', 'New Adult', 'Young Adult', 'Science', 'Poetry', 'Paranormal', 'Art', 'Psychology', 'Autobiography', 'Parenting', 'Adult Fiction', 'Humor', 'Horror', 'History', 'Food and Drink', 'Christian Fiction', 'Business', 'Biography', 'Thriller', 'Contemporary', 'Spirituality', 'Academic', 'Self', 'Help', 'Historical', 'Christian', 'Suspense', 'Short', 'Stories', 'Novels', 'Health', 'Politics', 'Cultural', 'Erotica', 'Crime']
```

Obtención de todos los enlaces referidos a las categorías de los libros

```
catalogo=[]
for enlaces in soup.find_all('a'):
  catalogo.append(enlaces.get('href'))
catalogo
```

```
filtrado=[]
for links in catalogo:
  if 'catalogue/category/books/' in links:
    filtrado.append("http://books.toscrape.com/" + links)
filtrado
```

```
# Con esto corroboramos que existen 50 categorías de libros
len(filtrado)
```

```
# Con esto probamos la visión que tenemos de una página principal de cada catgoría
page = requests.get("http://books.toscrape.com/catalogue/category/books/travel_2/index.html")
soup = BeautifulSoup(page.content)
print(soup.prettify())

```

```
# Se utiliza esta función para buscar la última barra de la direcicón y modificarla por lo que tenemos
# Esto es debido a que en el contenido de la página aparece cortado.
def findLastIndex(str, x): 
    index = -1
    for i in range(0, len(str)): 
        if str[i] == x: 
            index = i 
    return index 
  # https://www.geeksforgeeks.org/find-last-index-character-string/
```

### Obtención de todas las páginas de cada una de las categorías
En este apartado se tuvo en cuenta la opción "Next" que tiene cada página

```
import time
import random
filtradoxd = filtrado
for i in filtradoxd:
  time.sleep(random.uniform(0,10))
  page = requests.get(i)
  soup = BeautifulSoup(page.content)
  valor = findLastIndex(i,"/")
  print(i)
  if soup.find('li',class_='next') is not None:
    filtradoxd.append(i[0:valor+1] + soup.find('li',class_='next').find('a').get('href'))
  else:
    continue
print(filtradoxd)
```

```
filtradoxd.sort()
print(filtradoxd)
```

```
# Visualización de todas las páginas de las categorías disponibles ordenadas alfabéticamente
filtradoxd
```

### Obtención de todos los productos de todas las páginas de cada una de las categorías

```
productos = []

for enlace in filtradoxd:
  time.sleep(random.uniform(0,10))
  page = requests.get(enlace)
  soup = BeautifulSoup(page.content)
  for texto in soup.find_all("div",class_="image_container"):
    productos.append(texto.find('a').get('href').replace("../../../", "http://books.toscrape.com/catalogue/"))

productos
```

```
# Realizando esta operación corroboramos la existencia de los 1000 productos existentes en la propia página
len(productos)
```

```
# Inicialización de las listas de almacenamiento de cada uno de los libros
ltitle = []
limage = []
lrating = []
ldescription = []
lcategory = []
lUPC = []
lproducttype = []
lpriceextax = []
lpriceincltax = []
ltax = []
lavailability = []
lnumberreviews = []
```

### Obtención de la información de cada uno de los productos.

```
for i in productos:
  time.sleep(random.uniform(0,10))
  page = requests.get(i)
  soup = BeautifulSoup(page.content)
  ltitle.append(soup.find('h1').get_text())
  limage.append(soup.find('img').get('src').replace("../../", "http://books.toscrape.com/"))
  lrating.append(soup.find('p',class_="star-rating")['class'][1])
  ldescription.append(soup.find('h2').find_next().get_text())
  lcategory.append(soup.find("ul", class_="breadcrumb").find_all('a')[2].get_text())
  lUPC.append(soup.find("table", class_="table table-striped").find_all('td')[0].get_text())
  lproducttype.append(soup.find("table", class_="table table-striped").find_all('td')[1].get_text())
  lpriceextax.append(soup.find("table", class_="table table-striped").find_all('td')[2].get_text())
  lpriceincltax.append(soup.find("table", class_="table table-striped").find_all('td')[3].get_text())
  ltax.append(soup.find("table", class_="table table-striped").find_all('td')[4].get_text())
  lavailability.append(soup.find("table", class_="table table-striped").find_all('td')[5].get_text())
  lnumberreviews.append(soup.find("table", class_="table table-striped").find_all('td')[6].get_text())
```

### Descarga de todas las imágenes de los productos disponibles en el propia Drive por si se necesitaran

```
for i in limage:
  time.sleep(random.uniform(0,10))
  img_data = requests.get(i).content
  valor = findLastIndex(i,"/")
  i = i[valor+1:len(i)]
  ruta = "/content/drive/My Drive/images/" + i
  with open(ruta, 'wb') as handler:
    handler.write(img_data)
```

```
#https://stackoverflow.com/questions/53468558/adding-image-to-pandas-dataframe/53469293

# Almacenamiento de toda la información relativa a los libros en un dataframe, para
# almacenarlos en un ficheor HTML y CSV
import pandas as pd
from IPython.core.display import HTML
# intialise data of lists.
datos = {'Title': ltitle, 'ProductType': lproducttype, 'Category':lcategory,'Rating':lrating, 
                         'Product Description':ldescription, 'UPC':lUPC,'Price Excl. Tax':lpriceextax, 
                         'Price Incl.Tax':lpriceincltax, 'Tax':ltax, 'Avaliability':lavailability, 'Number of Reviews':lnumberreviews}




# Create DataFrame
df = pd.DataFrame(datos)

df['image'] = limage
# convert your links to html tags 
def path_to_image_html(path):
    return '<img src="'+ path + '" width="60" >'

pd.set_option('display.max_colwidth', -1)

pd.set_option('display.max_columns', None)
# Print the output.
print(df.head())
print(len(df))
```
